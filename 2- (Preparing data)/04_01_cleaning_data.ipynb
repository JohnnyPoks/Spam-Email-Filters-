```python
import pandas as pd
pd.set_option('display.max_colwidth', 100)

data = pd.read_csv("../Data/SMSSpamCollection.txt", sep='\t', header=None)
data.columns = ['label', 'content']

data.head()
```

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    result = "".join([char for char in text if char not in string.punctuation])
    return result

data['Clean_Content'] = data['content'].apply(lambda x: remove_punctuation(x))

data.head()
```

```python
import re
def tokennizer(text):
    words = re.split(r'\W+', text)
    return words

data['Tokennized_Content'] = data['Clean_Content'].apply(lambda x: tokennizer(x.lower()))

data.head()
```

```python
import nltk
# nltk.download('stopwords')
```

```python
en_stopWords = nltk.corpus.stopwords.words('english')
```

```python
en_stopWords
```

```python
def remove_stopwords(text):
    result = [word for word in text if word not in en_stopWords]
    return result

data['Stopwords_Removed_Content'] = data['Tokennized_Content'].apply(lambda x: remove_stopwords(x))

data.head()
```
